## 2) Data Wrangling

```python
PS4ds_Publisher = PS4ds.drop(columns = 'Year').groupby(['Publisher']).sum()
PS4ds_Publisher
```

```python
xBoxds_Publisher = xBoxds.drop(columns = 'Year').drop(columns = 'Pos').groupby(['Publisher']).sum()
xBoxds_Publisher
```

```python
xBoxds_Publisher.sort_values(by = 'Global', ascending = False)
```

```python
PS4ds_Publisher.sort_values(by = 'Global', ascending = False)
```

```python
xBoxds_Publisher = xBoxds_Publisher[xBoxds_Publisher.Global!= 0]
PS4ds_Publisher = PS4ds_Publisher[PS4ds_Publisher.Global!= 0]
```

## 3) Plotting Graphs

**Bar Plot of Sales in Millions vs Publishers for PS4**

```python
PS = sns.barplot(data = PS4ds_Publisher)
PS.set(xlabel="Publishers", ylabel = "Sales in millions")
plt.title("PS4 Sales by Publishers")
```

**Bar Plot of Sales in Millions vs Publishers for xBox**

```python
xbox = sns.barplot(data = xBoxds_Publisher)
xbox.set(xlabel="Publisher", ylabel = "Sales in millions")
plt.title("xBox Sales by Publisher")
```

**Line Plot of Sales in Millions vs Publishers for xBox**

```python
az = sns.lineplot(data=xBoxds_Publisher, linewidth=0.75)
az.set(xlabel="Publishers", ylabel = "Sales in millions")
plt.title("xBox Total Sales by Publishers by Region")
for item in az.get_xticklabels():
    item.set_rotation(90)
```

**Line Plot of Sales in Millions vs Publishers for PS4**

```python
az = sns.lineplot(data=PS4ds_Publisher, linewidth=0.75)
az.set(xlabel="Publishers", ylabel = "Sales in millions")
plt.title("PS4 Total Sales by Publishers by Region")
for item in az.get_xticklabels():
    item.set_rotation(90)
```

**Violin Plot of Sales in Millions vs Region for xBox**

```python
az = sns.violinplot(data=xBoxds_Publisher)
az.set(xlabel="Regions", ylabel = "Sales in millions")
plt.title("xBox Total Sales by Publishers by Region")
```

**Violon Plot of Sales in Millions vs Region for PS4**

```python
az = sns.violinplot(data=PS4ds_Publisher)
az.set(xlabel="Regions", ylabel = "Sales in millions")
plt.title("PS4 Total Sales by Publishers by Region")
```
